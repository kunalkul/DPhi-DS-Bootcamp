```python
def aboutdf (df):
    count_null = df.isnull().sum()
    df_stats = pd.DataFrame(index = df.columns, data =
                           {'datatype': df.dtypes,
                            'unique_values': df.nunique(),
                            'have_null?': df.isnull().any(),
                            'Number of null values' : count_null,
                            'percentage of null values' : count_null/df.shape[0]*100 })
    return df_stats
```

# Modeling

```python
# Separate input features and output feature
X = cd4.drop(columns = ['income_level'])       # input features
Y = cd4.income_level      # output feature
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

## Balancing data

## Method 1 - Upsampling

```python
from sklearn.utils import resample
```

```python
# concatenate training data back together
train_data = pd.concat([X_train, Y_train], axis = 1)
```

```python
# separate minority and majority class
less_50k = train_data[train_data.income_level==0]
more_50k = train_data[train_data.income_level==1]
```

```python
# Unsample minority; we are oversampling the minority class to match the number of majority classs
more_upsampled = resample(more_50k,
                           replace = True, # Sample with replacement
                           n_samples = len(less_50k), # Match number in majority class
                           random_state=27)
```

```python
# combine majority and upsampled minority
upsampled = pd.concat([less_50k, more_upsampled])
```

```python
# Now let's check the classes count
upsampled.income_level.value_counts()
```

## Method 2 - Downsampling

```python
# we are still using our separated class i.e. fraud and not_fraud from above
# Again we are removing the observations of the majority class to mathch the number of minority class
# downsample majority
less_downsampled = resample(less_50k,
                                replace = False, # sample without replacement
                                n_samples = len(more_50k), # match minority n
                                random_state = 27)
```

```python
# combine minority and downsampled majority
downsampled = pd.concat([less_downsampled, more_50k])    # Concatenation
```

```python
# let's check the classes counts
downsampled.income_level.value_counts()
```

## Method 3 - SMOTE

```python
# import SMOTE 
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state = 25, ratio = 1.0)   # again we are eqalizing both the classes
```

```python
# fit the sampling
X_train, Y_train = sm.fit_sample(X_train, Y_train)
```

```python
np.unique(Y_train, return_counts=True)     # Y_train is numpy array, so unique() functions returns the count of all the unique elements in the array
```

## split again

```python
# Separate input features and output feature
X_u = upsampled.drop(columns = ['income_level'])       # input features
Y_u = upsampled.income_level      # output feature
```

```python
# Separate input features and output feature
X_d = downsampled.drop(columns = ['income_level'])       # input features
Y_d = downsampled.income_level      # output feature
```

```python
X_train_u, X_test_u, Y_train_u, Y_test_u = train_test_split(X_u, Y_u, test_size = 0.3, random_state = 42)
```

```python
X_train_d, X_test_d, Y_train_d, Y_test_d = train_test_split(X_d, Y_d, test_size = 0.3, random_state = 42)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score
```

### 1) DecisionTreeClassifier

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train_u, Y_train_u)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train_d, Y_train_d)
```

```python
dtc_pred = dtc.predict(X_test_d)

accuracy_dtc = accuracy_score(Y_test_d, dtc_pred)
auc_roc_dtc = roc_auc_score(Y_test_d, dtc_pred)
f1_dtc = f1_score(Y_test_d, dtc_pred)
print("Accuracy score of the DTC model is: {}".format(accuracy_dtc))
print("roc_auc_score of the DTC model is: {}".format(auc_roc_dtc))
print("F1 score of the DTC model is: {}".format(f1_dtc))
```
